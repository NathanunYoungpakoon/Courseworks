\begin{align*}
P(spring) &= P(in\_stock, spring) + P(sold\_out, spring) \\
 &= 0.20 + 0.13 \\
 &=  0.33 \\
\end{align*}

\begin{align*}
P(sold\_out \ | \ fall) &= \frac{P(sold\_out,fall)}{P(fall)} \\
 &= \frac{0.23}{0.33} \\
 &=  0.697
\end{align*}
